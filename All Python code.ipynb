```python
import numpy as np
arr = np.array([[0, 1], [2, 3]])
print("Original flattened array:")
print(arr)
flat_arr = arr.flatten()
max_value = flat_arr.max()
min_value = flat_arr.min()
print("Maximum value of the above flattened array:", max_value)
print("Minimum value of the above flattened array:", min_value)
```

```python
import numpy as np
A = np.array([[5, 5],
              [1, 2]])

B = np.array([[0, 1],
              [1, 0]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)
result = np.dot(A, B)

print("\nResultant Matrix after multiplication:")
print(result)
transpose_result = result.T

print("\nTranspose of the resultant matrix:")
print(transpose_result)
```

```python
import numpy as np
A = np.array([
    [6, 1, 1, 3],
    [4, 2, 5, 1],
    [2, 8, 7, 6],
    [3, 1, 9, 7]
], dtype=float)
print("Original matrix A:")
print(A)
det = np.linalg.det(A)
print("\nDeterminant of A:", det)
if abs(det) < 1e-12:
    print("Matrix is singular (determinant â‰ˆ 0). Inverse does not exist.")
else:
    A_inv = np.linalg.inv(A)
    np.set_printoptions(precision=8, suppress=True)
    print("\nInverse of A (A^-1):")
    print(A_inv)
```
